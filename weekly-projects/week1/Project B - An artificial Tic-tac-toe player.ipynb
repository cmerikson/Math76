```python
import random
import sys

def print_board(board, move_number):
    print(f" \033[1mMove {move_number}\033[0m")
    for i, row in enumerate(board):
        print(" | ".join(row))
        if i < len(board) - 1:  # Avoid printing the bottom line of dashes
            print("-" * 10)
        print(' ')
    print("")

def check_winner(board, mark):
    win_conditions = [
        [board[0][0], board[0][1], board[0][2]],
        [board[1][0], board[1][1], board[1][2]],
        [board[2][0], board[2][1], board[2][2]],
        [board[0][0], board[1][0], board[2][0]],
        [board[0][1], board[1][1], board[2][1]],
        [board[0][2], board[1][2], board[2][2]],
        [board[0][0], board[1][1], board[2][2]],
        [board[2][0], board[1][1], board[0][2]],
    ]
    return [mark, mark, mark] in win_conditions

def get_empty_positions(board):
    positions = []
    for i in range(3):
        for j in range(3):
            if board[i][j] == " ":
                positions.append((i, j))
    return positions

def user_move(board):
    while True:
        try:
            row = input("Enter the row (1-3) or 'quit' to exit: ")
            if row.lower() == 'quit':
                print("Game ended by user.")
                sys.exit()
            row = int(row) - 1
            col = input("Enter the column (1-3) or 'quit' to exit: ")
            if col.lower() == 'quit':
                print("Game ended by user.")
                sys.exit()
            col = int(col) - 1
            if board[row][col] == " ":
                board[row][col] = "X"
                break
            else:
                print("This position is already taken.")
        except (ValueError, IndexError):
            print("Invalid input. Please enter numbers between 1 and 3.")


def empty_corner_test(empty_positions):
    corner_spots = [(0,0),(0,2),(2,0),(2,2)]
    empty_corners = []
    for i in corner_spots:
        if i in empty_positions:
            empty_corners.append(i)
    return empty_corners   

def center_test(empty_positions):
    center_spot = [(1,1)]
    center = []
    for i in center_spot:
        if i in empty_positions:
            center.append(i)   
    return center   

def find_opposite(board):
    if board[0][0] == "O" and board[2][2] == " ":
            opposite = [(2,2)]
    elif board[2][0] == "O" and board[0][2] == " ":
            opposite = [(0,2)]
    elif board[0][2] == "O" and board[2][0] == " ":
            opposite = [(2,0)]
    elif board[2][0] == "O" and board[0][0] == " ":
            opposite = [(0,0)]
    elif board[2][2] == "O" and board[0][0] == " ":
            opposite = [(0,0)]
    elif board[0][0] == "O" and board[0][2] == " ":
            opposite = [(0,2)]
    return opposite      

def find_counter(board):
    if board[1][1] == "O" and board[0][0] == "O":
            if board[2][2] == "X" and board[1][0] == "X":
                    counter = [(0,2)]
            elif board[2][2] == "X" and board[2][1] == "X":
                    counter = [(2,0)]
            elif board[2][2] == "X" and board[1][2] == "X":
                    counter = [(0,2)]
            elif board[2][2] == "X" and board[0][1] == "X":
                    counter = [(2,0)]
    elif board[1][1] == "O" and board[2][2] == "O":
            if board[0][0] == "X" and board[1][0] == "X":
                    counter = [(2,0)]
            elif board[0][0] == "X" and board[2][1] == "X":
                    counter = [(0,2)]
            elif board[0][0] == "X" and board[1][2] == "X":
                    counter = [(0,2)]
            elif board[0][0] == "X" and board[0][1] == "X":
                    counter = [(0,2)]
    elif board[1][1] == "O" and board[2][0] == "O":
            if board[0][2] == "X" and board[1][0] == "X":
                    counter = [(2,2)]
            elif board[0][2] == "X" and board[2][1] == "X":
                    counter = [(0,0)]
            elif board[0][2] == "X" and board[1][2] == "X":
                    counter = [(2,2)]
            elif board[0][2] == "X" and board[0][1] == "X":
                    counter = [(0,0)]
    elif board[1][1] == "O" and board[0][2] == "O":
            if board[2][0] == "X" and board[1][0] == "X":
                    counter = [(0,0)]
            elif board[2][0] == "X" and board[2][1] == "X":
                    counter = [(2,2)]
            elif board[2][0] == "X" and board[1][2] == "X":
                    counter = [(0,0)]
            elif board[2][0] == "X" and board[0][1] == "X":
                    counter = [(2,2)]
    else:
        counter = []
    return counter   

def seek_win(board, string):
    if board[0][0] == string and board[1][0] == string and board[2][0] == " ":
        end = [(2,0)]
    elif board[1][0] == string and board[2][0] == string and board[0][0] == " ":
        end = [(0,0)]
    elif board[0][0] == string and board[2][0] == string and board[1][0] == " ":
        end = [(1,0)]
    elif board[0][1] == string and board[1][1] == string and board[2][1] == " ":
        end = [(2,1)]
    elif board[1][1] == string and board[2][1] == string and board[0][1] == " ":
        end = [(0,1)]
    elif board[0][1] == string and board[2][1] == string and board[1][1] == " ":
        end = [(1,1)]
    elif board[0][2] == string and board[1][2] == string and board[2][2] == " ":
        end = [(2,2)]
    elif board[1][2] == string and board[2][2] == string and board[0][2] == " ":
        end = [(0,2)]
    elif board[0][2] == string and board[2][2] == string and board[1][2] == " ":
        end = [(1,2)]
    elif board[0][0] == string and board[1][1] == string and board[2][2] == " ":
        end = [(2,2)]
    elif board[0][0] == string and board[2][2] == string and board[1][1] == " ":
        end = [(1,1)]
    elif board[1][1] == string and board[2][2] == string and board[0][0] == " ":
        end = [(0,0)]
    elif board[1][2] == string and board[1][1] == string and board[2][0] == " ":
        end = [(2,0)]
    elif board[2][0] == string and board[1][1] == string and board[0][2] == " ":
        end = [(0,2)]
    elif board[2][0] == string and board[0][2] == string and board[1][1] == " ":
        end = [(1,1)]
    elif board[0][0] == string and board[0][1] == string and board[0][2] == " ":
        end = [(0,2)]
    elif board[0][0] == string and board[0][2] == string and board[0][1] == " ":
        end = [(0,1)]   
    elif board[0][2] == string and board[0][1] == string and board[0][0] == " ":
        end = [(0,0)]
    elif board[1][0] == string and board[1][1] == string and board[1][2] == " ":
        end = [(1,2)]
    elif board[1][0] == string and board[1][2] == string and board[1][1] == " ":
        end = [(1,1)]   
    elif board[1][2] == string and board[1][1] == string and board[1][0] == " ":
        end = [(1,0)]
    elif board[2][0] == string and board[2][1] == string and board[2][2] == " ":
        end = [(2,2)]
    elif board[2][0] == string and board[2][2] == string and board[2][1] == " ":
        end = [(2,1)]   
    elif board[2][2] == string and board[2][1] == string and board[2][0] == " ":
        end = [(2,0)]
    elif board[2][0] == string and board[1][1] == string and board[0][2] == " ":
        end = [(0,2)]
    elif board[0][2] == string and board[1][1] == string and board[2][0] == " ":
        end = [(2,0)]
    elif board[0][2] == string and board[2][0] == string and board[1][1] == " ":
        end = [(1,1)]
    else:
        end = []
    return end

def computer_move(board, move_number):
    empty_positions = get_empty_positions(board)
    if move_number+1 < 3:
        empty_corners = empty_corner_test(empty_positions)
        if len(empty_corners) == 4:
            move = random.choice(empty_corners)
        else:
            move = (1,1)   
    elif move_number+1 == 3:
        empty_center = center_test(empty_positions)
        if len(empty_center) == 1:
            move = (1,1)  
        else:
            opposite = find_opposite(board)
            move = opposite[0]
    elif move_number+1 > 3:
        possible_win = seek_win(board, "O")
        if len(possible_win) == 1:
            move = possible_win[0]
        elif len(possible_win) == 0:
            possible_loss = seek_win(board, "X")
            if len(possible_loss) == 1:
                move = possible_loss[0]
            else:
                counter = find_counter(board)
                if len(counter) == 1:
                    move = counter[0]
                else:
                    move = random.choice(empty_positions)
    board[move[0]][move[1]] = "O"

def tic_tac_toe():
    board = [[" " for _ in range(3)] for _ in range(3)]
    print("Welcome to Tic Tac Toe.")
    user_first = input("Do you want to go first? (y/n) or 'quit' to exit: ").lower()
    if user_first == 'quit':
        print("Game ended by user.")
        return
    user_first = user_first == 'y'
    
    for move_number in range(9):
        print_board(board, move_number)
        if user_first:
            user_move(board)
            if check_winner(board, "X"):
                print_board(board, move_number)
                print("Congratulations, you have won.")
                return
            user_first = False
        else:
            computer_move(board, move_number)
            if check_winner(board, "O"):
                print_board(board, move_number)
                print("The computer wins.")
                return
            user_first = True
        
        if not get_empty_positions(board):
            print("Draw")
            return

if __name__ == "__main__":
    tic_tac_toe()
```
